# **Feature Encoding**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df.head(7)
```

```python
df.info()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.Gender.value_counts()
```

```python
## Handling missing values
df.Gender.mode()
```

```python
df.Gender = df.Gender.fillna('Male')
```

```python
df.Gender.value_counts()
```

# **Label Encoding**

```python
le = preprocessing.LabelEncoder()
df['Gender_label'] = le.fit_transform(df.Gender.values)
```

```python
df.head(7)
```

```python
df.Gender_label.value_counts()
```

# **One Hot Encoding**

```python
df_dummies = pd.get_dummies(df)

df_dummies.head(7)
```

```python
df1.head()
```

```python
df1.Gender_Encoded.value_counts()
```

####Benifits of Tagret Encoding
- It is simple and quick encoding meathod that doesn't add to the dimensionality of the dataset.

####Limitations of TE
- It is dependent on the target distribution which means target encoding requires careful validation as it can be prone to overfitting.

```python
X = df.Gender
y = df.Exited
```

```python
ce_hash = HashingEncoder(cols=['Gender'])
ce_hash.fit_transform(X,y)
```

```python
X = df.Geography
y = df.Exited

ce_hash = HashingEncoder(cols=['Geography'])
ce_hash.fit_transform(X,y)
```
